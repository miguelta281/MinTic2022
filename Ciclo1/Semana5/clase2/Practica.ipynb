```python
#Extraer una columna
SeriesEdades = data.Edad

#Contar repeticiones (personas por edad)
SeriesEdades.value_counts().plot(kind = 'bar') #Agregar el .plot() para mostrar las gr√°ficas (visualizar rapidamente)
```
